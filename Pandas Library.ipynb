# Slicing of DataFrame
- loc or iloc function is used for accessing certain data from DataFrame
- loc function accepts column names while iloc function accepts column names
- loc function includes last index and iloc function excludes last index

```python
# Exctract Records from 30-35 from symboling to aspiration column by using loc function
data.loc[4:8,'symboling':'aspiration']
```

```python
# Extract only 5th and 16th row from aspiration and symboling column by using  loc function
data.loc[[5,16],['symboling','aspiration']]
```

```python
# Exctract Records from 30-35 from symboling to aspiration column by using loc function
data.iloc[30:36,3:6]
```

```python
# Extract only 5th and 16th row from aspiration and symboling column by using  loc function
data.iloc[[5,16],[0,4]]
```

# Pandas Filtering and Groupby

```python
import pandas as pd
df = pd.read_csv('Salaries.csv')
df
```

```python
data2.head()
```

```python
# Extract record where gender = Female
df['gender'] == 'Female'
```

```python
# we doesn't want values like true or false, we want all table values according to Female.
# so for that we made silly change in above program, we just add square brackets
```

```python
# masking
df[df['gender'] =='Female']
```

```python
df[df['gender'] == 'female']
```

```python
#Gender is female and salary is greater than 150000
df[(df['gender'] == 'Female') & (df['salary'] > 150000)]
```

```python
df[(df['gender'] == 'Female') | (df['salary'] > 150000)]
```

```python
#What is the average salary of rank prof?
df[df['rank'] == 'Prof']
```

```python
df[df['rank'] == 'Prof']['salary'].mean()
```

```python
#Average salary of prof rank and gender is Male
df[(df['gender'] == 'Male') & (df['rank'] == 'Prof')]['salary'].mean()
```

```python
# find Average salary of every unique rank and gender male.
print(df[(df['rank'] == 'Prof') & (df['gender'] == 'Male')]['salary'].mean())
print(df[(df['rank'] == 'AsstProf') & (df['gender'] == 'Male')]['salary'].mean())
print(df[(df['rank'] == 'AssocProf') & (df['gender'] == 'Male')]['salary'].mean())
```

- if column have 100 unique values then we have to write this program 100 times but this is not infeasible program
- so for avoiding this we use 'Groupby' function 

# Groupby aggregate function

```python
df.groupby('rank')['salary'].agg({'mean','median','min','max'})
```
